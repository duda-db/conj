# questão um:

```python
import matplotlib.pyplot as plt

R = 10
S = 2
T = 8 * S / R * 10**6
T = T / 60
print(f"Tempo necessário para baixar um arquivo de {S} GB com taxa de transferência de {R} Mbps é {T:.2f} minutos!")
lista1 = list(range(1, 21))
lista2 = [8 * S / R * 10**6 / 60 for S in lista1]

plt.plot(lista1, lista2)
plt.title("Tempo de download vs tamanho de arquivo")
plt.xlabel("GB")
plt.ylabel("minutos")
plt.show()
```

```python
# PRIMEIRA FUNÇÃO

import math
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return np.exp(x) - 4*x**2

x = np.linspace(-2, 2, 1000)
y = f(x)
plt.plot(x, y, label=r'$e^x - 4x^2 = 0$')
plt.ylabel('Função decrescente')
plt.title('$f(x) = e^x - 4x^2 sendo y = 0$')

plt.legend()
plt.show()

```

```python
# SEGUNDA FUNÇÃO

import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return np.log(3*(x**2+1)) - np.log(3*x) - 2

x_val = np.linspace(0.1, 10, 1000)
y_val = func(x_val)

plt.plot(x_val, y_val)
plt.title('Gráfico da função log3(x²+1) - log3(x) = 2')
plt.ylabel('Função crescente')

plt.show()

```

```python
#1 

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(2*x)/(x**2)


x = np.linspace(-10, 10, 1000)
y = f(x)


plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = \frac{\sin(2x)}{x^2}$')
plt.show()

```

```python
#2

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x')
f = sp.exp(x) / x**3
f_num = sp.lambdify(x, f)

xs = np.linspace(0, 5, 1000)
ys = f_num(xs)


plt.plot(xs, ys)
plt.xlim([0, 5])
plt.ylim([0, 50])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = \frac{e^x}{x^3}$')
plt.show()
```

passo a passo do #1, aplicando a regra de L'Hôpital:

$$\lim_{x\to 0} \frac{\sin(2x)}{x} \cdot \frac{1}{x}$$
$$= \lim_{x\to 0} \frac{2\cos(2x)}{1}$$
$$= 2$$


passo a passo do #2, aplicando a regra de L'Hôpital tambem:

$\lim_{x \to \infty} \frac{e^x}{x^3} = lim_{x \to \infty} \frac{e^x}{3x^2}$

$\lim_{x \to \infty} \frac{e^x}{3x^2} = lim_{x \to \infty} \frac{e^x}{6x}$

$\lim_{x \to \infty} \frac{e^x}{6x} = lim_{x \to \infty} \frac{e^x}{6} = \infty /6 = \infty$
