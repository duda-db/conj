# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt

f1 = lambda x: x * np.exp(-x**2)
f2 = lambda x: x + 1

x_intersect = np.roots([1, -1, -1])

xvalores = np.linspace(0, 2, 100)
y1valores = f1(xvalores)
y2valores = f2(xvalores)


plt.fill_between(xvalores, y1valores, y2valores, where=(y1valores > y2valores), color='grey', alpha=0.3)
plt.plot(xvalores, y1valores, label='y = xe^(-x^2)',color='pink')
plt.plot(xvalores, y2valores, label='y = x + 1' ,color='magenta')
plt.xlim(0, 2)
plt.ylim(0, max(max(y1valores), max(y2valores)) + 1)

plt.legend()
plt.show()





```

a)
Para calcularmos o valor da integral ∫(20x^2 + 1) dx, antes precisamos encontrar a antiderivada da função integranda e, em seguida, aplicar os limites de integração.

```python
#a)
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 20*x**2 + 1

a = -2
b = 2

def F(x):
    return (20/3)*x**3 + x
integral = F(b) - F(a)
print("O valor da integral definida é:", integral)


x = np.linspace(-3, 3, 100)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='20x^2 + 1',color='grey')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('gráfico da função integranda')
plt.axvline(x=a, color='pink', linestyle='--', label='limite inferior de integração')
plt.axvline(x=b, color='purple', linestyle='--', label='limite superior de integração')
plt.legend()
plt.grid(True)
plt.show()

```

Primeiro precisamos encontrar a antiderivada da função f(x). A antiderivada da função f(x) é a função F(x) que, ao derivá-la, resulta em f(x). Para isso, integramos termo a termo:


$\int (3x^3 - 2x+1)dx = \int 3x^3 dx - \int 2x  dx + \int 1 dx $



Colocando a regra de integração para cada termo:

$F(x) = \frac{3}{4}x^4 - x^2 + x + C$

Aplicamos a condição F(1) = 0 para encontrar o valor de C.
Substituindo x por 1 na função F(x) e igualando a zero:


$F(1) = \frac{3}{4}1^4 - 1^2 + 1 + C = 0$


Simplificando a expressão:


$ \frac{3}{4} - 1 + 1 + C = 0$


$ \frac{3}{4} + C = 0$


Subtraimos 3/4 de ambos os lados da equação:


$ C = -\frac{3}{4}$

Portanto, o valor da constante C para a antiderivada F(x) da função $f(x) = 3x^3 - 2x + 1$

de modo que F(1) = 0, e C  $ = -\frac{3}{4}$.
